```python
counter=0
def actfact(n):
    global counter
    counter+=1
    print(counter)
    if (n==0 or n==1):
        return 1
    else:
        return(n*actfact(n-1))
```

```python
def fibanocci(n):
    if (n==0):
        return 0
    elif(n==1):
        return 1
    return fibanocci(n-1)+fibanocci(n-2)
```
